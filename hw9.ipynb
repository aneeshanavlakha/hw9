##### By Aneesha Navlakha

Natural Language Processing, or NLP, is a massive buzzword that everyone seems to be throwing into their portfolios/job descriptions/marketing strategies. #### Background
Natural Language Processing (NLP) is a branch of ML that aims to help computers comprehend, analyze, and produce. Basically, given that humans speak human and computers speak computer, we give computers the tools to understand human enough to perform operations on it. Sort of like giving a child shape-matching games where they have a bunch of random shape blocks and need to learn the difference between the shapes in order to use them. 

Say we add new complex shapes into the child's toy collection. Once it has learned details about each shape individually, it can start to draw higher level inferences on all the shapes as a group - it can diferentiate between dissimilar shapes, like a circle and a square, or match common shapes and group them into clusters like squares and rectangles in one group (4 sides) and triangles in another (3 sides). In other words, NLP is useful in many ways, one being to find the common themes in a large corpus of documents, especially when the corpus is too large to manually sift through. In my assignments, my goal was to find the common themes in a large corpus of documents - the "20 newsgroups text dataset" - using scikit-learn's LatentDirichletAllocation (LDA) NLP model. This process of discovering high-level themes in a large collection of texts is called "Topic modeling".

As my assignment said, atypical to a regular corpus, the 20 newsgroups text dataset is already labeled with themes (e.g., sports, hardware, religion), which I could use to cross-check the topics discovered by my models. The themes I used are 'alt.atheism', 'comp.graphics', 'comp.windows.x',  'rec.sport.baseball', 'rec.sport.hockey', 'soc.religion.christian', 'talk.politics.guns', and 'talk.politics.mideast'. #### Preprocessing
These are a few of the sentences in the corpus (Figure 2)

```python
Image("img/Figure 2.png")
```

As you can see, the sentences in this corpus make little sense, with a lot of unnecessary text, characters, and spaces. It makes the sentences harder to work with so before I handed the corpus over the the LDA model, I cleaned it up in a process called "pre-processing". For this, I used SpaCy, a popular NLP library that is commonly used to pre-process text. I defined a function that did the following:

- replaced newlines/tabs with spaces
- cleaned the text by removing irrelevant tokens
- normalized text by replacing URLs, emails, and numbers with placeholders
- simplified words by lemmatizing them
- ensured consistency through lower-casing
- output the cleaned and processed text as a string

Note: "Lemmatization" is a process that analyzes a word's context and morphology to return the root form (lemma), ensuring the base form is a valid word in the language, e.g., the lemma of both "carrying" and "carried" is "carry". "Word tokenization" splits sentences into words.

The text_pp column in Figure 3 below shows the cleaned up text.

```python
Image("img/Figure 3.png")
```

#### Topic Modelling
Finally, I passed my pre-processed text into the LDA model to perform topic modeling! As a reminder, topic modeling is the process of discovering high-level themes in a large collection of texts. The text was now clean enough to work with i.e., the splinters and dust had been removed from the shapes so now we could start to find themes. The sentences were first converted into individual words to get the most important information.

LDA allowed me to explore and identify the ideal number of themes to sort the words into. Using a range of values for the the n_components hyper-parameter in LDA, I tried values ranging from 1 - 50. For some reason, however, the model kept recommending the largest number so I picked 7 after inspecting the contents of topics at each value in the next step. This is a potential shortcoming that I will discuss at the end of my article. Figure 4 below shows how many words were in each of the 7 topics (ignore the fancy axis names).

#### Word Topic Association

This process showed me what the prominent words in each cluster are. The features of the different toys, e.g., how many sides, has been extracted and we have defined the appropriate number of clusters based on that.

#### Document Topic Association
But does each document only contain one theme or are there some which are a mixture of multiple themes? Can a shape with 2 sets of parallel sides be both a quadrilateral, and a rectangle?

The document_topic[] function of LDA gives you a percentage of how similar each document in your corpus is to each theme (topic) you defined. I found the document_topic for the first 5 documents (sentences) i.e., document_topics[:5], in our original table . This is what I got:

- Document 0 corresponds to Topic 0 the most at 66% and Topic 1 at 32%
- Document 1 corresponds to Topic 2 the most at 74%
- Document 2 corresponds to Topic 1 the most at 61% and Topic 2 at 32%
- Document 3 corresponds to Topic 2 the most at 70%
- Document 4 corresponds to Topic 0 the most at 85%

There are only some differences from the pattern of assignments that the corpus had by default: document 2 that was earlier classified as comp.graphics is now mostly religion and belief. The remaining documents seem to be classified along the same lines.

This process showed me that there are documents which are a mixture of multiple themes.

#### Learnings
